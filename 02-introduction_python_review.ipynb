## A Quick Tour of Python Language Syntax

- Python’s syntax simplicity
- Python’s syntax structure and features


## Comments

- Use # for single-line comments
- No multi-line comment syntax
- Inline comments example: x += 2  # shorthand for x = x + 2


## Statement Termination

- End of line terminates statements
- Continuation with \ or parentheses
- Example: x = (1 + 2 + 3 + 4 + 5 + 6 + 7 + 8)


## Optional Semicolons

- Use semicolons for multiple statements on one line
- Example: lower = []; upper = []
- Generally discouraged by style guides


## Whitespace and Indentation

- Indentation denotes code blocks, unlike C/C++
- Consistent indentation is required
- Indentation example with for loops and if statements


## Whitespace Within Lines

- Whitespace within lines does not affect execution
- Aids readability, especially with operators
- Examples of equal expressions with different spacing


## Parentheses Usage

- For grouping statements or function calls
- Function calls require parentheses, even with no arguments
- Example: L.sort()


## Basic Python Semantics: Variables and Objects

- Focus on Python semantics: meaning of statements
- Variables and objects as core data structures


## Variables as Pointers

- Variables are pointers, not containers
- No need for variable type declaration
- Dynamic typing: variable can point to any object type


## Example: Variables as Pointers

- x = 1, x = 'hello', x = [1, 2, 3] (dynamic typing)
- Python is dynamically-typed, unlike C’s static typing


## Mutable Objects and Pointers

- Variables point to mutable objects
- Changing one variable affects others pointing to the same object
- Example with lists: x.append(4)


## Reassignment and Immutability

- Reassignment changes what variables point to
- Immutable types: numbers, strings, etc.
- Example: x += 5 doesn’t affect y


## Everything Is an Object

- Python is object-oriented: everything is an object
- Types are associated with objects, not variable names
- Example: type(x)


## Attributes and Methods

- Objects have metadata (attributes) and functionality (methods)
- Dot syntax: L.append(100), x.is_integer()
- Even simple types have methods


## Example: Methods in Python

- Access methods with dot and parentheses: x.is_integer()
- x.real, x.imag for numerical types
- type(x.is_integer) demonstrates everything is an object


## Basic Python Semantics: Operators

- Overview of semantics for Python operators
- Tools for comparing and operating on data


## Arithmetic Operations

- Seven binary arithmetic operators: +, -, *, /, //, %, **
- True division vs Floor division (/ vs //)
- Unary operators: - (negation), + (unary plus)


## Example: Arithmetic Operations

- Addition, subtraction, multiplication examples
- True division: 11 / 2 vs Floor division: 11 // 2
- Matrix multiplication operator: @ (introduced in Python 3.5)


## Bitwise Operations

- Six bitwise operators: &, |, ^, <<, >>, ~
- Works on binary representations of integers
- Example: bin(), bitwise operations on 4 and 10


## Assignment Operations

- Variable assignment with =
- Combined operation and assignment: a += 2
- Augmented assignments for arithmetic and bitwise operations


## Comparison Operations

- Comparison operators: ==, !=, <, >, <=, >=
- Returns Boolean values True or False
- Chained comparisons for complex tests


## Boolean Operations

- Boolean operators: and, or, not
- Used for combinatory truth or falsehood expressions
- Example: (x < 6) and (x > 2)


## Identity and Membership Operators

- Identity operators: is, is not for object identity
- Membership operators: in, not in for element presence
- Examples: checking identical objects, membership in lists


## Built-In Types: Simple Values

- Overview of Python’s basic scalar types
- Contrast with compound types in the next section


## Python Scalar Types

- int: integers
- float: floating-point numbers
- complex: complex numbers
- bool: Boolean values
- str: strings
- NoneType: special None object


## Integers

- Whole numbers without a decimal
- Variable-precision: no overflow like C integers
- Division upcasts to float: 5 / 2 = 2.5


## Floating-Point Numbers

- Store fractional numbers
- Standard and exponential notations: 1.4e6
- Limited precision: 0.1 + 0.2 != 0.3


## Complex Numbers

- Real and imaginary parts
- Use j suffix for imaginary parts: 1 + 2j
- Methods: .real, .imag, .conjugate(), abs()


## String Type

- Created with single or double quotes
- Useful methods: .upper(), .capitalize(), concatenation, repetition
- Indexing: zero-based, e.g., message[0]


## None Type

- Special type with single value: None
- Default return value for functions with no return
- Example: type(None)


## Boolean Type

- Two possible values: True, False
- Result of comparison operations
- Convertible from other types: bool()

- Built-in string methods and formatting
- Introduction to regular expressions for pattern matching


## Simple String Manipulation

- Define strings with single or double quotes

## String Space Formatting

## Substring Operations

- Methods return new strings


## Splitting and Joining Strings

- Split strings with .split(), .splitlines()
- Partition strings with .partition(), .rpartition()
- Join strings with "".join(iterable)


## Format Strings

- Format codes for numerical values

## Built-In Data Structures

- Compound types act as containers for other types
- Python’s built-in compound types


## Types Overview

- list: Ordered, mutable collection
- tuple: Ordered, immutable collection
- dict: Unordered, key-value mapping
- set: Unordered collection of unique values


## Lists

- Ordered and mutable
- Defined with square brackets, e.g., [2, 3, 5]
- Common operations: len(), .append(), concatenation, .sort()


## List Indexing and Slicing

- Access and modify elements using square brackets
- Zero-based indexing, negative indices for reverse order
- Slicing to access sub-lists


## Tuples

- Ordered and immutable
- Defined with parentheses or no punctuation, e.g., (1, 2, 3)
- Operations similar to lists, but immutable


## Dictionaries

- Flexible key-value mappings
- Access with keys, e.g., {'one': 1}
- Fast access due to internal hash table implementation


## Sets

- Unordered collections of unique items
- Defined with curly brackets, e.g., {2, 3, 5}
- Support for set operations: union, intersection, difference


## More Specialized Data Structures

- Found in collections module
- Named tuples, defaultdict, OrderedDict
- Extend functionality of standard types


## List Comprehensions

- Efficient and readable way to create lists
- Syntax: [expr for var in iterable]


## Basic List Comprehensions

- Compress a list-building loop into one line
- Example: list of square numbers [n ** 2 for n in range(12)]


## Multiple Iteration

- Use multiple for expressions for cross-product
- Example: [(i, j) for i in range(2) for j in range(3)]


## Conditionals on the Iterator

- Filter items with a condition
- Example: exclude multiples of 3 [val for val in range(20) if val % 3 > 0]


## Conditionals on the Value

- Use if-else conditionals in comprehensions
- Example: [val if val % 2 else -val for val in range(20) if val % 3]


## Set and Dict Comprehensions

- Use {} for sets: {n**2 for n in range(12)}
- Use {key: value} for dicts: {n: n**2 for n in range(6)}

# Temperature converter 

Complete the following program that converts temperature from degree F to degree C

The output for the example is left on the notebook for your reference

The formual for conversion is $ C = (F − 32) × 5/9  $

```python
print("Enter temperature in degree F:")
temp_f = input()
```

```python
temp_f = float(temp_f)
temp_c = (temp_f - 32) * 5 / 9
print(f"The temprature in degree C is {temp_c:.2f}")
```

# Temperature converter in English

Modify the above program to accept in put in English, for example instead of `66`, the user enters `sixty six`

Assume that the valid range of values is zero to ninety nine

```python
print("Enter temperature in degree F:")
temp_f = input()
```

```python
#hint use this dict
letter2number = {'zero':0, 'one':1, 'two':2, 'three':3, 'four':4, 'five':5, 'six':6, 'seven':7, 'eight':8, 'nine':9,
                'ten':10, 'twenty':20, 'thirty':30, 'forty':40, 'fifty':50, 'sixty':60, 'seventy':70, 'eighty':80, 'ninety':90}
```

```python
comps = temp_f.lower().split()
temp_f = sum( letter2number[comp] for comp in comps )
temp_c = (temp_f - 32) * 5 / 9
print(f"The temprature in degree C is {temp_c:.2f}")    
```

Select records of products from China

```python
# sample records of products, each product has a name, price, and country of origin
records = [{'name': 'office chair', 'price': 150, 'coo':'China'},
           {'name': 'bike', 'price': 800, 'coo': 'China'},
           {'name': 'car', 'price': 40000, 'coo': 'Germany'},
           {'name': 'laptop', 'price': 2000, 'coo': 'USA'},
           {'name': 'desk', 'price': 300, 'coo': 'China'},
           {'name': 'lamp', 'price': 50, 'coo': 'China'}]
```

```python
[record for record in records if record['coo'] == 'China']
```

What is the most expensive product?

```python
highest_price = max(record['price'] for record in records)
[record for record in records if record['price'] == highest_price][0]
```

# WTOP.3 — Basic Python Syntax

Python uses **indentation** to delimit code blocks (instead of braces).
Statements end at the **end of a line** (no semicolon required).
`#` begins a comment; there is no multi-line comment syntax.

```python
# Comments: everything after # is ignored
x = 10        # inline comment

# Long expressions can be continued with \ or inside parentheses
total = (1 + 2 + 3
         + 4 + 5)   # parentheses allow line continuation
print(total)
```

```python
# Indentation defines blocks — be consistent (4 spaces is standard)
for i in range(3):
    if i % 2 == 0:
        print(i, "is even")
    else:
        print(i, "is odd")
```

# WTOP.4 — Python Semantics: Variables and Objects

In Python, **variables are pointers**, not containers.
- No type declaration needed — Python is *dynamically typed*.
- Multiple variables can point to the same object.
- **Mutable** objects (lists, dicts) can be changed in place.
- **Immutable** objects (int, str, tuple) cannot be changed; reassignment creates a new object.
- Everything in Python is an **object** with attributes and methods.

```python
# Dynamic typing — same variable, different types
x = 42
print(type(x))
x = "hello"
print(type(x))
x = [1, 2, 3]
print(type(x))
```

```python
# Variables as pointers: x and y point to the SAME list
x = [1, 2, 3]
y = x          # y points to the same object, not a copy
x.append(4)
print(y)       # [1, 2, 3, 4] — y sees the change!

# To copy, use list() or slicing
z = list(x)
x.append(5)
print(z)       # [1, 2, 3, 4] — z is unaffected
```

```python
# Reassignment vs. mutation
x = 10
y = x
x += 5          # integers are immutable — x now points to a new object
print(x, y)     # 15 10 — y is unaffected

# Everything is an object with attributes and methods
n = 3.14
print(n.is_integer())   # float method
print(n.real, n.imag)   # numeric attributes
```

# WTOP.5 — Python Semantics: Operators

| Category | Operators |
|----------|-----------|
| Arithmetic | `+  -  *  /  //  %  **` |
| Comparison | `==  !=  <  >  <=  >=` |
| Boolean | `and  or  not` |
| Bitwise | `&  |  ^  <<  >>  ~` |
| Assignment | `=  +=  -=  *=  /=  //=  **=  %=` |
| Identity | `is  is not` |
| Membership | `in  not in` |

`/` is true division (always returns float); `//` is floor division.

```python
# Arithmetic
print(11 / 2)    # 5.5  — true division
print(11 // 2)   # 5    — floor division
print(11 % 2)    # 1    — remainder
print(2 ** 10)   # 1024 — exponentiation
```

```python
# Comparison — return True or False; can be chained
x = 5
print(x > 2)
print(1 < x < 10)   # chained: equivalent to (1 < x) and (x < 10)

# Boolean operators
print((x > 2) and (x < 10))
print((x < 0) or (x > 3))
print(not (x == 5))
```

```python
# Identity vs. equality
a = [1, 2, 3]
b = [1, 2, 3]
c = a

print(a == b)   # True  — same content
print(a is b)   # False — different objects
print(a is c)   # True  — same object

# Membership
print(2 in a)
print(5 not in a)
```

# WTOP.6 — Built-In Scalar Types

| Type | Example | Notes |
|------|---------|-------|
| `int` | `42` | Arbitrary precision — no overflow |
| `float` | `3.14`, `1.4e6` | IEEE 754 double precision; limited precision |
| `complex` | `1 + 2j` | `.real`, `.imag`, `.conjugate()` |
| `str` | `'hello'` | Immutable sequence of characters |
| `bool` | `True`, `False` | Subclass of int (True==1, False==0) |
| `NoneType` | `None` | Singleton; default function return value |

```python
# int — arbitrary precision
print(2 ** 100)

# float — limited precision
print(0.1 + 0.2)          # not exactly 0.3!
print(0.1 + 0.2 == 0.3)   # False

# complex
c = 2 + 3j
print(c.real, c.imag)
print(abs(c))              # magnitude
```

```python
# str basics
s = "Hello, World!"
print(len(s))
print(s.upper())
print(s[0])       # indexing
print(s[-1])      # last character
print(s[0:5])     # slicing

# f-strings (Python 3.6+)
name = "Alice"
score = 95.678
print(f"Student: {name}, Score: {score:.1f}")
```

```python
# bool and None
print(bool(0), bool(1), bool(""), bool("hello"), bool([]))
x = None
print(x is None)       # canonical way to check for None
print(type(None))
```

# WTOP.7 — Built-In Data Structures

| Type | Ordered | Mutable | Unique values | Syntax |
|------|---------|---------|---------------|--------|
| `list` | ✓ | ✓ | | `[1, 2, 3]` |
| `tuple` | ✓ | ✗ | | `(1, 2, 3)` |
| `dict` | ✓* | ✓ | keys only | `{'a': 1}` |
| `set` | ✗ | ✓ | ✓ | `{1, 2, 3}` |

*dicts preserve insertion order since Python 3.7

```python
# list — ordered, mutable
primes = [2, 3, 5, 7, 11]
primes.append(13)
primes[0] = 1          # mutable
print(primes)
print(primes[1:4])     # slicing returns a new list
print(primes[-1])      # last element
```

```python
# tuple — ordered, immutable (good for fixed records)
point = (3.0, 4.0)
x, y = point           # unpacking
print(x, y)
# point[0] = 99       # TypeError — tuples are immutable

# Common use: returning multiple values from a function
def min_max(lst):
    return min(lst), max(lst)

lo, hi = min_max([4, 1, 9, 2])
print(lo, hi)
```

```python
# dict — key-value mapping
person = {'name': 'Alice', 'age': 30, 'city': 'Athens'}
print(person['name'])
person['age'] = 31          # update
person['email'] = 'a@b.com' # add new key
print(person.keys())
print(person.values())
print(person.get('phone', 'N/A'))  # safe access with default
```

```python
# set — unordered, unique items
a = {1, 2, 3, 4}
b = {3, 4, 5, 6}
print(a | b)    # union
print(a & b)    # intersection
print(a - b)    # difference
print(2 in a)   # membership test is O(1)

# remove duplicates from a list
words = ['apple', 'banana', 'apple', 'cherry', 'banana']
print(list(set(words)))
```

# WTOP.8 — Control Flow

- `if / elif / else` — conditional execution
- `for` — iterate over any iterable
- `while` — loop until a condition is False
- `break` — exit the loop immediately
- `continue` — skip to the next iteration
- `else` on a loop — runs only if the loop completed without `break`

```python
# if / elif / else
def classify(n):
    if n < 0:
        return "negative"
    elif n == 0:
        return "zero"
    else:
        return "positive"

for n in [-3, 0, 7]:
    print(n, "->", classify(n))
```

```python
# for loop — iterates over any sequence
fruits = ['apple', 'banana', 'cherry']
for fruit in fruits:
    print(fruit.upper())

# range() generates integers
for i in range(0, 10, 2):   # start, stop, step
    print(i, end=' ')
print()
```

```python
# enumerate and zip — essential loop patterns
names = ['Alice', 'Bob', 'Carol']
scores = [92, 87, 95]

for i, name in enumerate(names):
    print(f"  {i}: {name}")

for name, score in zip(names, scores):
    print(f"  {name}: {score}")
```

```python
# while, break, continue
# Print odd numbers < 10 using continue
i = 0
while i < 10:
    i += 1
    if i % 2 == 0:
        continue        # skip even
    print(i, end=' ')
print()

# break: first number divisible by 7 in range 1..100
for n in range(1, 101):
    if n % 7 == 0:
        print("First multiple of 7:", n)
        break
```

```python
# Loop else: runs only if no break occurred
# Classic use: Sieve step — check if n has a divisor
def is_prime(n):
    for i in range(2, n):
        if n % i == 0:
            break
    else:
        return True
    return False

primes = [n for n in range(2, 30) if is_prime(n)]
print(primes)
```

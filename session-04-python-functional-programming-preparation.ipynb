# Functional Programming in Python

Functional programming (FP) is a paradigm in which computation proceeds primarily through the evaluation of **pure functions** — functions whose output depends solely on their inputs and that produce no observable side effects (Real Python, n.d.).

Python is not a purely functional language, but it supports the FP style through first-class functions, `lambda` expressions, and the built-ins `map()`, `filter()`, and `reduce()` (Mertz, 2015).

FP code tends to be:
- **High-level**: describe *what* you want, not *how* to compute it.
- **Transparent**: a pure function's behaviour is fully described by its inputs and outputs.
- **Parallelizable**: side-effect-free routines can run concurrently.

# Example 1: First-Class Functions

In Python, **functions are first-class citizens**: they can be assigned to variables, passed as arguments to other functions, and returned as return values (Real Python, n.d.).

A function that takes or returns another function is called a **higher-order function (HOF)** (Mertz, 2015).

````python
# 1a. Assign a function to a variable
def square(x):
    return x ** 2

sq = square        # pass the function object — no parentheses
print(sq(5))       # 25

# 1b. Pass a function as an argument
def apply(func, value):
    return func(value)

print(apply(square, 7))   # 49

# 1c. Return a function from a function (factory / closure)
def make_multiplier(n):
    def multiplier(x):
        return x * n
    return multiplier

triple = make_multiplier(3)
print(triple(10))        # 30

# 1d. sorted() uses a key= callback — a standard-library HOF
animals = ["ferret", "vole", "dog", "gecko"]
print(sorted(animals))                       # lexicographic order
print(sorted(animals, key=len))              # by string length
print(sorted(animals, key=len, reverse=True))  # longest first
````

# Example 2: Lambda Expressions

A `lambda` creates an **anonymous function** inline. Syntax:

```
lambda <parameters>: <expression>
```

The body must be a single expression; no statements, no `return` keyword (Real Python, n.d.). Lambda expressions shine when you need a short, throwaway function as an argument to a HOF.

````python
# 2a. Basic lambda
reverse = lambda s: s[::-1]
print(reverse("hello"))    # olleh

# 2b. Immediately-invoked lambda
print((lambda x, y: x + y)(3, 4))    # 7

# 2c. Conditional expression inside lambda
parity = lambda x: "even" if x % 2 == 0 else "odd"
print(parity(7))    # odd

# 2d. Lambda as the key= argument to sorted()
words = ["banana", "kiwi", "apple", "fig", "cherry"]
print(sorted(words, key=lambda s: -len(s)))   # longest first

# Note: if you need to reuse the function, define it with "def" instead
# of assigning a name to a lambda (Real Python, n.d.).
````

# Example 3: map() — Transform Every Element

`map(f, iterable)` applies `f` to each element and returns a **lazy iterator** of the results (Real Python, n.d.). Wrap it in `list()` to materialise the values.

Because `map()` is lazy, it is memory-efficient for large sequences — the equivalent of Java's `Stream.map()`.

````python
# 3a. map() with a named function
def celsius_to_fahrenheit(c):
    return c * 9 / 5 + 32

temps_c = [0, 20, 37, 100]
temps_f = list(map(celsius_to_fahrenheit, temps_c))
print(temps_f)    # [32.0, 68.0, 98.6, 212.0]

# 3b. map() with a lambda (more concise for simple transforms)
print(list(map(lambda c: c * 9 / 5 + 32, temps_c)))

# 3c. Practical: convert numbers to strings for str.join()
nums = [1, 2, 3, 4, 5]
print("+".join(map(str, nums)))   # "1+2+3+4+5"

# 3d. Comprehension equivalent (often preferred in Python, per Real Python, n.d.)
print([c * 9 / 5 + 32 for c in temps_c])
````

# Example 4: map() with Multiple Iterables

Pass *n* iterables to `map()` and it applies the function to the corresponding elements of each iterable in parallel. The function must accept *n* arguments (Real Python, n.d.).

````python
prices     = [10.0, 25.0, 5.50]
quantities = [3,    2,    10]

# Multiply corresponding elements from two lists
line_totals = list(map(lambda p, q: p * q, prices, quantities))
print(line_totals)    # [30.0, 50.0, 55.0]

# Three iterables — element-wise sum
a = [1, 2, 3]
b = [10, 20, 30]
c = [100, 200, 300]
print(list(map(lambda x, y, z: x + y + z, a, b, c)))   # [111, 222, 333]

# Comprehension + zip equivalent
print([p * q for p, q in zip(prices, quantities)])
````

# Example 5: filter() — Select Elements

`filter(f, iterable)` returns a lazy iterator of all elements for which `f` is **truthy**; elements where `f` is falsy are discarded (Real Python, n.d.).

This is Python's equivalent of Java's `Stream.filter()`.

````python
# 5a. Filter even numbers
evens = list(filter(lambda x: x % 2 == 0, range(10)))
print(evens)    # [0, 2, 4, 6, 8]

# 5b. Filter strings by property
animals = ["cat", "Cat", "CAT", "dog", "Dog", "DOG"]
print(list(filter(lambda s: s.isupper(), animals)))   # ["CAT", "DOG"]

# 5c. Remove falsy values: empty strings are falsy, so filter(lambda s: s, ...)
#     keeps only non-empty strings
mixed = ["", "hello", "", "world", ""]
print(list(filter(lambda s: s, mixed)))   # ["hello", "world"]

# 5d. Comprehension equivalent
print([x for x in range(10) if x % 2 == 0])
````

# Example 6: reduce() — Fold to a Single Value

`reduce(f, iterable)` applies `f` to the first two elements, then to that result and the third element, and so on, **collapsing the iterable to a single value** (Real Python, n.d.).

Guido van Rossum disliked `reduce()` and removed it from the built-ins; import it from `functools`. Its most natural uses involve associative operators (+, *, etc.) (Real Python, n.d.).

As Mertz (2015) observes, `map()` and `filter()` are actually special cases of `reduce()` — `reduce()` is the most general of the three primitives.

````python
from functools import reduce

# 6a. Sum via reduce  (sum() is simpler, but reduce illustrates the pattern)
total = reduce(lambda x, y: x + y, [1, 2, 3, 4, 5])
print(total)    # 15

# 6b. Factorial  (reduce over a range)
def factorial(n):
    return reduce(lambda x, y: x * y, range(1, n + 1))

print(factorial(5))    # 120

# 6c. Maximum via reduce
nums = [23, 49, 6, 32]
print(reduce(lambda x, y: x if x > y else y, nums))   # 49

# 6d. String concatenation via reduce
words = ["cat", "dog", "hedgehog", "gecko"]
print(reduce(lambda x, y: x + y, words))   # "catdoghedgehoggecko"
# Note: "".join(words) is more Pythonic (Real Python, n.d.)
````

# Example 7: reduce() with an Initializer

An optional third argument supplies the **initial accumulator value**. The first call to `f` receives `(initializer, first_element)` instead of `(first_element, second_element)` (Real Python, n.d.).

Use an initializer when the iterable may be empty, or when you need a non-default starting point (e.g., building a dict or list).

````python
from functools import reduce

# 7a. Sum starting from 100
print(reduce(lambda x, y: x + y, [1, 2, 3, 4, 5], 100))   # 115

# 7b. Build a dict from (key, value) pairs
pairs = [("a", 1), ("b", 2), ("c", 3)]
result = reduce(lambda acc, kv: {**acc, kv[0]: kv[1]}, pairs, {})
print(result)   # {"a": 1, "b": 2, "c": 3}

# 7c. Demonstrate that map() is a special case of reduce() (Mertz, 2015)
add5 = lambda n: n + 5
via_reduce = reduce(lambda acc, x: acc + [add5(x)], range(5), [])
via_map    = list(map(add5, range(5)))
print(via_reduce)   # [5, 6, 7, 8, 9]
print(via_map)      # [5, 6, 7, 8, 9]
````

# Example 8: The operator Module

Every Python infix operator has a named function equivalent in the `operator` module. These named functions are faster than equivalent lambdas and read more naturally when passed to higher-order functions (Mertz, 2015).

````python
import operator
from functools import reduce

data = [1, 2, 3, 4, 5]

# 8a. operator.add vs lambda
sum_lambda   = reduce(lambda a, b: a + b, data, 0)
sum_operator = reduce(operator.add,        data, 0)
print(sum_lambda, sum_operator)    # 15  15

# 8b. operator.mul for factorial
product = reduce(operator.mul, data, 1)
print(product)    # 120

# 8c. operator.itemgetter as a sort key (works like a lambda r: r["key"])
records = [
    {"name": "Bob",   "score": 82},
    {"name": "Alice", "score": 95},
    {"name": "Carol", "score": 78},
]
ranked = sorted(records, key=operator.itemgetter("score"), reverse=True)
for r in ranked:
    print(f"  {r['name']:8s}  {r['score']}")
````

# Example 9: functools.partial() — Partial Application

`partial(f, *pre_args)` returns a new function with some arguments of `f` pre-filled, reducing the arity of the original function. This is Python's spelling of **currying** (Mertz, 2015).

`partial()` is particularly useful for creating specialised predicates to pass to `filter()` or `map()`.

````python
from functools import partial

# 9a. Specialise a general function
def power(base, exp):
    return base ** exp

square = partial(power, exp=2)
cube   = partial(power, exp=3)

print(square(5))    # 25
print(cube(3))      # 27

# 9b. Composable predicates with partial + operator (Mertz, 2015)
is_lt100 = partial(operator.gt, 100)   # operator.gt(100, x)  ->  100 > x
is_gt10  = partial(operator.lt, 10)    # operator.lt(10, x)   ->  10 < x
print(is_lt100(50), is_gt10(50))   # True  True

# 9c. Department-specific filter built from a general predicate
def in_dept(dept, record):
    return record["dept"] == dept

is_engineering = partial(in_dept, "Engineering")

people = [
    {"name": "Alice", "dept": "Engineering"},
    {"name": "Bob",   "dept": "Marketing"},
    {"name": "Carol", "dept": "Engineering"},
]
print([p["name"] for p in filter(is_engineering, people)])
# ["Alice", "Carol"]
````

# Example 10: FP Style vs Pythonic Style

Guido van Rossum advocated replacing `map()` and `filter()` with comprehensions, which most Python developers find more readable (Real Python, n.d.).

| Task | FP style | Pythonic equivalent |
|---|---|---|
| Transform | `list(map(f, xs))` | `[f(x) for x in xs]` |
| Select | `list(filter(p, xs))` | `[x for x in xs if p(x)]` |
| Both | `list(map(f, filter(p, xs)))` | `[f(x) for x in xs if p(x)]` |

Use whichever is clearer. The FP style shines when composing **named functions** rather than inline lambdas, or when building lazy pipelines over large iterables.

````python
nums = range(1, 11)

# Square only the even numbers — two styles
fp_style = list(map(lambda x: x**2, filter(lambda x: x % 2 == 0, nums)))
pythonic  = [x**2 for x in nums if x % 2 == 0]

print(fp_style)    # [4, 16, 36, 64, 100]
print(pythonic)    # [4, 16, 36, 64, 100]

# With named functions the FP style reads cleanly as a pipeline:
is_even  = lambda x: x % 2 == 0
squared  = lambda x: x ** 2

result = list(map(squared, filter(is_even, nums)))
print(result)   # [4, 16, 36, 64, 100]
````
